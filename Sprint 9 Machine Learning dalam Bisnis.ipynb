## 3.1 Model Regresi Linier

```python
# Fungsi untuk membagi dataset menjadi training set dan validation set dengan rasio 75:25
def pembagi_dataset(data):
    features = data.drop(columns=['product', 'id'])
    target =  data["product"]
    features_train, features_test, target_train, target_test = train_test_split(features, target, test_size=0.25, random_state=12345)
    print("features_train:",len(features_train),"\ntarget_train:",len(target_train),"\nfeatures_test:",len(features_test),"\ntarget_test:",len(target_test))
    return features_train, features_test, target_train, target_test

# Fungsi normalisasi fitur dan melatih model dengan regresi linier
def linear_regression(features_train, features_test, target_train, target_test):
    to_normalize = ['f0','f1','f2']
    scaler = StandardScaler()
    scaler.fit(features_train[to_normalize])
    features_train.loc[:,to_normalize] = scaler.transform(features_train[to_normalize])
    features_test.loc[:,to_normalize] = scaler.transform(features_test[to_normalize])
    lr = LinearRegression().fit(features_train,target_train)
    predicted_values = lr.predict(features_test)
    print('R2:', round(lr.score(features_test, target_test),2))
    print('RMSE:', round(np.sqrt(mean_squared_error(target_test, predicted_values)),2))
    print('Volume Rata-Rata:',round(predicted_values.mean(),2))
    return predicted_values
```

```python
# Menjalankan Fungsi untuk menampilkan ukuran pembagian data dan prediksi model
print()
print("\033[1m" + 'Wilayah I' + "\033[0m")
print()
print("\033[4mRasio Fitur dan Target (75:25)\033[0m")
features_train0, features_test0, target_train0, target_test0 = pembagi_dataset(ga)
print()
print("\033[4mPrediksi\033[0m")
predicted_values0 = linear_regression(features_train0, features_test0, target_train0, target_test0)
print()
print("\033[1m" + 'Wilayah II' + "\033[0m")
print()
print("\033[4mUkuran Fitur dan Target (75:25)\033[0m")
features_train1, features_test1, target_train1, target_test1 = pembagi_dataset(gb)
print()
print("\033[4mPrediksi\033[0m")
predicted_values1 = linear_regression(features_train1, features_test1, target_train1, target_test1)
print()
print("\033[1m" + 'Wilayah III' + "\033[0m")
print()
print("\033[4mUkuran Fitur dan Target (75:25)\033[0m")
features_train2, features_test2, target_train2, target_test2 = pembagi_dataset(gc)
print()
print("\033[4mPrediksi\033[0m")
predicted_values2 = linear_regression(features_train2, features_test2, target_train2, target_test2)
```

## 6. Perhitungan Laba dan Resiko

Pada bagian ini dibuat fungsi untuk membuat 1000 sample dari 500 sumur acak dengan teknik bootstrapping. Tujuan fungsi ini adalah untuk mencari distribusi laba.

```python
#Fungsi teknik bootstrapping untuk menemukan distribusi laba
def bootstrapping_laba(target, prediksi):
    daftar_laba = []
    target = pd.Series(target).reset_index(drop=True)
    state = np.random.RandomState(12345)
    for i in range(1000):
        target_subsample = target.sample(n=titik, replace=True, random_state = state)
        prediksi_subsample = prediksi[target_subsample.index]
        daftar_laba.append(laba(target_subsample, prediksi_subsample))

    daftar_laba = pd.Series(daftar_laba)
    print('Titik yang menghasilkan laba:',daftar_laba.gt(0).sum())
    print('Persentase titik yang menghasilkan laba:', round((daftar_laba.gt(0).sum()/len(daftar_laba))*100,2,),'%')
    return daftar_laba
```

Hitunglah risiko dan laba untuk setiap wilayah:
1. Gunakan teknik bootstrapping dengan 1.000 sampel untuk menemukan distribusi labanya.

**2. Laba rata-rata, selang kepercayaan, dan risiko kerugian.**

Fungsi berikut ini dibuat untuk perhitungan:
1. Laba rata-rata di setiap wilayah.
2. Resiko kerugian tidak boleh lebih dari 2.5% sehingga selang kepercayaan yang digunakan adalah 95%.
3. Persentase kerugian.
4. Interval atas dan interval bawah dari selang kepercayaan

```python
#Membuat fungsi perhitungan laba rata-rata, selang kepercayaan dan persentase kerugian
def selang_kepercayaan(laba_wilayah):
    ci = st.t.interval(0.95, len(laba_wilayah)-1, laba_wilayah.mean(), laba_wilayah.sem())
    resiko = (len(laba_wilayah[laba_wilayah < 0]) / len(laba_wilayah))* 100
    bawah = laba_wilayah.quantile(0.025)
    atas = laba_wilayah.quantile(0.975)
    print('Laba Rata-rata pada wilayah ini adalah: ', round(laba_wilayah.mean(),2))
    print('Selang kepercayaan 95% adalah:', ci[0], ',' ,ci[1] )
    print('Resiko kerugian adalah', resiko, '%')
    print('Interval kuartil kepercayaan', round(bawah,2), ',',round(atas,2))
```

```python
# Menampilkan laba rata-rata, selang kepercayaan dan persentase kerugian untuk setiap wilayah
print("\033[4mWilayah I:\033[0m")
selang_kepercayaan(w1)
print()
print("\033[4mWilayah II:\033[0m")
selang_kepercayaan(w2)
print()
print("\033[4mWilayah III:\033[0m")
selang_kepercayaan(w3)
```
